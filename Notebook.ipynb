```python
# The website dataset.puglia.it collects a list of information collected from the Apulian region.
# This notebook provides an analysis of absence percentage of employees using the dataset at the following link
# http://www.dataset.puglia.it/dataset/tassi-assenza-dipendenti
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Importing the dataframe to be analyzed

dataset = pd.read_csv('tassiassenzadipendenti.csv')
```

```python
# Reviewing the top lines of the dataset to give a first look at the data.

dataset.head()
```

```python
# Using the describe function to collect general information of the dataset

dataset.describe()
```

```python
# Assigning values of month and absences to variables

month = dataset['mese']
abs_days = dataset['tot_giornate_assenza']
```

```python
# Checking the two new partial dataframes

month.head()
```

```python
abs_days.head()
```

```python
# Reviewing absence rate per month

fig, ax = plt.subplots()
ax.bar(month, abs_days, align='center')
plt.title('Absence rate per month')
plt.ylabel('Days of absence')
plt.rcParams["figure.figsize"] = (15,5)
plt.grid()
```
